```python
import pandas as pd
```

```python
# load two dataframes and keep only rows that are in df2 but not in df1
df1 = pd.read_csv("../export.csv", on_bad_lines="skip", encoding="ISO-8859-1")
df2 = pd.read_csv("../merged_exports.csv", on_bad_lines="skip", encoding="ISO-8859-1")

df_merged = df2.merge(df1, indicator=True, how='left', on=['title', 'doi'])
# Keep only rows that exist in df2 but not in df1
df_diff = df_merged[df_merged['_merge'] == 'left_only'].drop('_merge', axis=1)

df_diff.to_csv("keep_snowballing.csv", index=False)
```
