```python
import pandas as pd
import json
import os
```

```python
# load all json files from folder tmp_store, then concatenate into a single dataframe
json_files = [f for f in os.listdir("tmp_store") if f.endswith(".json")]
dataframes = []
for f in json_files:
    with open(os.path.join("tmp_store", f), "r") as file:
        data = json.load(file)
        data = data["results"]
        df = pd.DataFrame(data)
        dataframes.append(df)
merged_df = pd.concat(dataframes, ignore_index=True)

merged_df.to_csv("merged_tmp_store.csv", index=False)
```

```python
coll_alex = pd.read_csv('tmp_store/openalex.csv')
coll_zotero = pd.read_csv('tmp_store/zotero.csv')
# rename column Title to title
coll_alex = coll_alex.rename(columns={'Title': 'title'})

# get difference between coll_alex and coll_zotero based on title
diff_df = coll_alex.merge(coll_zotero, indicator=True, how='left', on=['title'])
diff_df = diff_df[diff_df['_merge'] == 'left_only'].drop('_merge', axis=1)

diff_df
```
