```python
import pandas as pd
```

```python
# openalex and zotero difference comparison
df1 = pd.read_csv("openalex.csv", on_bad_lines="skip")
df2 = pd.read_csv("zotero.csv", on_bad_lines="skip")

# get columns title and doi
df1 = df1[['title', 'doi']]
df2.rename(columns={'Title': 'title', 'DOI': 'doi'}, inplace=True)
df2 = df2[['title', 'doi']]

# remove https://doi.org/ from doi column in both dataframes
df1['doi'] = df1['doi'].str.replace('https://doi.org/', '', regex=False)
# set all rows in doi column to lowercase
df1['doi'] = df1['doi'].str.lower()
df2['doi'] = df2['doi'].str.lower()

# concat df along title and doi columns
concat_df = pd.concat([df1, df2], ignore_index=True)
concat_df.drop_duplicates(inplace=True, keep=False, ignore_index=True)

concat_df


```

```python
# zotero datasets only difference
df1 = pd.read_csv("seed.csv")
df2 = pd.read_csv("snowballing_depth1.csv")
df3 = pd.read_csv("snowballing_depth2.csv")
df4 = pd.read_csv("snowballing_depth3.csv")

# concat along title and doi columns
concat_df = pd.concat([df1, df2, df3], ignore_index=True)

concat_df
```
