```python
import pandas as pd
```

```python
paper = pd.read_csv("13_10_25_screening_v3.csv", on_bad_lines="skip", encoding="ISO-8859-1")
paper = paper[paper["inclusion"] == 1]
paper.head()
```

```python
# get average scores
paper['average_score'] = paper[['score_clarity_research_design', 'score_methodological_rigor', 'score_validation_methods']].mean(axis=1)

paper.head()
```

```python
# get average across all papers
average_clarity = paper['score_clarity_research_design'].mean(axis=0)
average_rigor = paper['score_methodological_rigor'].mean(axis=0)
average_validation = paper['score_validation_methods'].mean(axis=0)
average_overall = paper['average_score'].mean(axis=0)

print(f"Average clarity score: {average_clarity}")
print(f"Average rigor score: {average_rigor}")
print(f"Average validation score: {average_validation}")
print(f"Average overall score: {average_overall}")
```

```python
paper_qa = paper[paper['average_score'] >= 1.0]
difference = len(paper) - len(paper_qa)
print(f"Dropped {difference} papers below quality threshold")
dropped_papers = paper[paper['average_score'] < 1.0]
print(f"\n{dropped_papers["ï»¿title"].to_string(index=False)}")
```

```python
# drop papers with any score below 1
paper_qa_strict = paper[(paper['score_clarity_research_design'] >= 1.0) &
                        (paper['score_methodological_rigor'] >= 1.0) &
                        (paper['score_validation_methods'] >= 1.0)]
difference_strict = len(paper) - len(paper_qa_strict)
print(f"Dropped {difference_strict} papers below strict quality threshold")
dropped_papers_strict = paper[~paper.index.isin(paper_qa_strict.index)]
print(f"\n{dropped_papers_strict["ï»¿title"].to_string(index=False)}")
```
