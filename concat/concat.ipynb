```python
import pandas as pd
import os
```

```python
csv_files = [f for f in os.listdir(".") if f.endswith(".csv")]

# read and concatenate all csv files
dataframes = [pd.read_csv(os.path.join("", f)) for f in csv_files]
concat = pd.concat(dataframes, ignore_index=True)
```

```python
concat.to_csv("concat.csv", index=False)
```
