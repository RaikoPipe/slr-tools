```python
import pandas as pd
```

```python
# walk through folder merge and find all csv files

import os

csv_files = [f for f in os.listdir("") if f.endswith(".csv")]

# read and concatenate all csv files
dataframes = [pd.read_csv(os.path.join("", f), on_bad_lines="skip", encoding="ISO-8859-1") for f in csv_files]
merged_df = pd.concat(dataframes, ignore_index=True)

merged_df.to_csv("merged_exports.csv", index=False)
```
